```python
import matplotlib.pyplot as plt

# Function to format currency
def fEUR(x): return f"EUR {x:,.0f}"

# ---- Cost Ranges (EUR) ----
ai_costs = {
    'Personnel': (1_400_000, 1_600_000),
    'Partnerships': (500_000, 700_000),
    'Infrastructure': (200_000, 300_000),
    'Hackathons': (50_000, 100_000),
    'Testing/Pilot': (800_000, 1_000_000),
    'Compliance': (150_000, 200_000),
    'Miscellaneous': (100_000, 400_000)
}

ar_costs = {
    'Personnel': (600_000, 800_000),
    'Partnerships': (300_000, 400_000),
    'Infrastructure': (100_000, 150_000),
    'Hackathons': (30_000, 50_000),
    'Testing/Pilot': (200_000, 300_000),
    'Compliance': (50_000, 80_000),
    'Miscellaneous': (220_000, 220_000)
}

def calculate_midpoint(cost_dict):
    mid = {cat: (low+high)/2 for cat, (low, high) in cost_dict.items()}
    base_cost = sum(mid.values())
    return mid, base_cost

ai_mid, ai_base = calculate_midpoint(ai_costs)
ar_mid, ar_base = calculate_midpoint(ar_costs)


ai_maint = ai_base * 0.20
ar_maint = ar_base * 0.20


def partner_costs(base, type_):
    if type_=="Big Companies":
        return base*1.2 + base*1.2*0.10*1.10
    elif type_=="Mid-Tier":
        return base*1.10
    elif type_=="Startups":
        return base*0.90 + base*0.90*0.10*0.90
    elif type_=="Young Talent":
        reduction = 0.175 if base>2_000_000 else 0.225
        return base*(1-reduction) + base*(1-reduction)*0.10*0.90

partner_types = ['Big Companies', 'Mid-Tier', 'Startups', 'Young Talent']
ai_partners = [partner_costs(ai_base, p) for p in partner_types]
ar_partners = [partner_costs(ar_base, p) for p in partner_types]

print("AI Personalization Ecosystem Base Cost:", fEUR(ai_base), " | Maintenance:", fEUR(ai_maint))
print("AR/VR Mind Oasis Base Cost:", fEUR(ar_base), " | Maintenance:", fEUR(ar_maint))

for i,p in enumerate(partner_types):
    print(f"{p} (AI): {fEUR(ai_partners[i])} | (AR/VR): {fEUR(ar_partners[i])}")
```

```python
# ---- Bar Chart ----
x = range(len(partner_types))
plt.figure(figsize=(10,6))
plt.bar([i-0.2 for i in x], ai_partners, width=0.4, label='AI Personalization', color='skyblue')
plt.bar([i+0.2 for i in x], ar_partners, width=0.4, label='AR/VR Mind Oasis', color='lightcoral')
plt.xticks(x, partner_types)
plt.ylabel('Total Cost (EUR)')
plt.title('Rituals Project Cost by Partner Type')
plt.legend()
plt.tight_layout()
plt.show()
```
